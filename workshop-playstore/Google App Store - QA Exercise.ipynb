```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('googleplaystore.csv') #"https://chukmunnlee.github.io/dataset/googleplaystore.csv")
```

```python
# Total No of Apps in each category
res=df.groupby("Category").size()
print(res)    
```

```python
#Avg app rating for each category
res=df.groupby("Category").mean()
res
```

```python
# Drop na
df.dropna(inplace=True)
```

```python
#Find category with the highest avg rating 
res=df.groupby("Category").mean()
res[res.Rating == df.Rating.max()]
```

```python
#Plot out diff ranges.. 
plt.figure(figsize=(7,7))
# 5 - 4.5 
High_range = df.loc[(df['Rating'] > 4.5) & (df['Rating'] <= 5.0 )]
High_Rating_Result=High_range.groupby("Rating").size()
High_Rating_Result.plot(kind="line", figsize=(20,10))
plt.show()

# 4.5 - 3.5
Mid_High_range = df.loc[(df['Rating'] > 3.5) & (df['Rating'] <= 4.5)]
Mid_High_Rating_Result=Mid_High_range.groupby("Rating").size()
Mid_High_Rating_Result.plot(kind="line", figsize=(20,10))
plt.show()

# 3.5 - 2.5

Mid_range = df.loc[(df['Rating'] > 2.5) & (df['Rating'] <= 3.5 )]
Mid_Rating_Result=Mid_range.groupby("Rating").size()
Mid_Rating_Result.plot(kind="line", figsize=(20,10))
plt.show()

# 2.5
Low_range = df.loc[(df['Rating'] == 2.5)]
Low_Rating_Result=Low_range.groupby("Rating").size()
Low_Rating_Result.plot(kind="bar", figsize=(20,10))
plt.show()
```

```python
#Plot out diff app rating (range). 
# 5 - 4.5 
# 4.5 - 3.5
# 3.5 - 2.5
# 2.5
arr = np.array([High_range.count(),Mid_High_range.count(),Mid_range.count(),Low_range.count()])
 
plt.plot(["5 - 4.5", "4.5 - 3.5","3.5 - 2.5", "2.5"],arr)
```
